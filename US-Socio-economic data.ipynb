### 2.	Data Preparation (5 marks)

a.	Scale / Transform/ clean the data so that it is suitable for model building.


```python
df=df.replace("?",np.NaN)
df.isnull().sum()
for i in df.columns:
    if df[i].isnull().sum()/df.shape[0]>0.80:
        df.drop(i,axis=1,inplace=True)
```

```python
for i in df.select_dtypes(include=np.number):
    df[i].fillna(df[i].median(),inplace=True)  #Median imputation in numeric columns for missing values
```

```python
for i in df.select_dtypes(include=np.object):
    df[i].fillna(df[i].mode()[0],inplace=True) #Mode imputation in numeric columns for missing values
```

```python
df.isnull().sum()
df.select_dtypes(include=np.object)
df["county"].value_counts()
def count(i):
    if i==17:
        return 2
    else:
        return 1
    
df["county"]=df["county"].astype(int).apply(count)
```

```python
df["community"]
def comm(i):
    if i==2060:
        return 2
    else:
        return 1
    
df["community"]=df["community"].astype(int).apply(comm)
```

```python
df.select_dtypes(include=np.object)
dic1=dict(zip(list(df["communityname"].value_counts().index),list(df["communityname"].value_counts().values)))
df["communityname"]=df["communityname"].map(dic1)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2,random_state=10)
pca.fit(df.drop("state",axis=1))
#a
# PCA can be applied when there is multicollinearity in columns.It reduces the multicollinearity due to 
#creation of orthogonal dimensions.
```

```python
sum(pca.explained_variance_ratio_)
dfvis=pd.DataFrame(pca.fit_transform(df.drop("state",axis=1)))
```

```python
#D
sns.scatterplot(dfvis[0],dfvis[1])  # Appears like there should be 7 different clusters in the data
```

```python
#c
pca=PCA(n_components=20,random_state=10)
pca_df=pd.DataFrame(pca.fit_transform(df.drop("state",axis=1)))
pca_df.head()
```

```python
for i in pca_df.columns:
    plt.tight_layout()
    sns.distplot(pca_df[i])
    plt.show()
    
#Spread of data PCA 
```

```python
for i in pca_df.columns:
    plt.tight_layout()
    sns.boxplot(pca_df[i])
    plt.show()  #Outliers present
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=[variance_inflation_factor(pca_df.values,i) for i in range(pca_df.shape[1])]
pd.DataFrame({"VIF":vif[:]},index=pca_df.columns)
#As can be seen, multicollinearity is less than 5 so multicollinearity is reduced due to orthogonalisation
#proerty of PCA
```

```python
#e

for i in pca_df.columns:
    q1=pca_df[i].quantile(0.25)
    q3=pca_df[i].quantile(0.75)
    iqr=q3-q1
    ul=q3+1.5*iqr
    ll=q1-1.5*iqr
    pca_df.loc[(pca_df[i]<ll),i]=ll
    pca_df.loc[(pca_df[i]>ul),i]=ul
    
#Ouliers have been treated
```

```python
wcss=[]
kvalues=[2,3,4,5,6,7,8,9,10]
from sklearn.cluster import KMeans
for k in kvalues:
    model=KMeans(n_clusters=k,random_state=10)
    model.fit(pca_df)
    wcss.append(model.inertia_)
    
plt.plot(kvalues,wcss)
#There is no visible elbow in the graph so we have to find optimal k value via silhpuette score
```

```python
from sklearn.metrics import silhouette_score
score=[]
for k in kvalues:
    model=KMeans(n_clusters=k,random_state=10)
    model.fit(pca_df)
    print("For k =",k,"the silhouette score is",silhouette_score(pca_df,model.labels_))
    
#Since k=2 has highest silhouette score   
```

```python
import yellowbrick
from yellowbrick.cluster import SilhouetteVisualizer
for k in kvalues:
    model=KMeans(n_clusters=k,random_state=10)
    vis=SilhouetteVisualizer(model,colors="yellowbrick")
    vis.fit(pca_df)
    vis.show()
    
#Optimum clusters are 2(NO outliers, above average values,relatively high silhouette score)
```

```python
model=KMeans(n_clusters=2,random_state=10)
model.fit(pca_df)
pca_df["clusters"]=model.labels_
```
